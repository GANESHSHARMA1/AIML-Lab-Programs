# 1. Apply multiclass classification (KNN) on suitable dataset.



```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.datasets import load_iris
from sklearn.metrics import classification_report, confusion_matrix

iris = load_iris()

X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.2)

print("Size of training data and It's label ", X_train.shape, y_train.shape)
print("Size of testing data and It's label ", X_test.shape, y_test.shape)
print("\n\n")

for i in range(len(iris.target_names)):
  print("Label ", i , " - ", iris.target_names[i])
print("\n\n")

classifier = KNeighborsClassifier(n_neighbors=2)
classifier.fit(X_train, y_train)

y_pred = classifier.predict(X_test)

print("Result of K-nn")

for i in range(0, len(X_test)):
  print("Sample : ", str(X_test[i]) , " Actual label : ", str(y_test[i]), " Predicted Label : ", str(y_pred[i]))
print("\n\n")

print("Classification Accuracy : ", classifier.score(X_test, y_test))
print("\n\n")
print("Confusion Matrix : ")
print(confusion_matrix(y_test, y_pred))
print("\n\n")
print("Accuracy Matrix : ")
print(classification_report(y_test, y_pred))
```

# 2. Apply PCA on the suitable dataset.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.datasets import load_breast_cancer

cancer = load_breast_cancer()

print("\n",cancer.keys())

print()
for i in range(len(cancer.target_names)):
  print("Label : ", i , " - ", cancer.target_names[i])
print()

# print("\n",cancer['DESCR'])

df = pd.DataFrame(cancer['data'], columns= cancer['feature_names'])

print(df.head())

scaler = StandardScaler()
scaler.fit(df)

scaled_data = scaler.transform(df)
print("\n", scaled_data.shape)

pca = PCA(n_components=2)
pca.fit(scaled_data)

x_pca = pca.transform(scaled_data)
print(x_pca.shape)

plt.figure(figsize=(8,6))
plt.scatter(x_pca[:,0], x_pca[:,1], c = cancer['target'])
plt.xlabel("PCA First Components")
plt.ylabel("PCA second Components")
plt.title("PCA Analysis")
plt.show()
```

# 5. Construct Backpropagation.



```python
import numpy as np

x = np.array([[2,9],[1,5],[3,6]], dtype=float)
y = np.array([[92],[86],[89]], dtype=float)
x = x / np.amax(x, axis=0)
y = y/100

epoch = 5000
lr = 0.1
input_neurons = 2
hidden_neurons = 3
output_neuron = 1

def sigmoid(x):
  return 1 / (1+np.exp(-x))
def derivatives_sigmoid(x):
  return x * (1-x)

wh = np.random.uniform(size = (input_neurons, hidden_neurons))
bh = np.random.uniform(size = (1, hidden_neurons))
wout = np.random.uniform(size=(hidden_neurons, output_neuron))
bout = np.random.uniform(size = (1, output_neuron))

for epoch in range(10):
  hinp1 = np.dot(x, wh)
  hinp = hinp1 + bh
  hlayer_act = sigmoid(hinp)
  outinp1 = np.dot(hlayer_act, wout)
  outinp = outinp1 + bout
  output = sigmoid(outinp)

  E0 = y - output
  outgrad = derivatives_sigmoid(output)
  d_output = E0 * outgrad
  EH = d_output.dot(wout.T)
  hiddengrad = derivatives_sigmoid(hlayer_act)
  d_hidden = EH * hiddengrad
  wout += hlayer_act.T.dot(d_output)*lr
  wh += x.T.dot(d_hidden)*lr

print("input :\n",x)
print("actual Output :\n",y)
print("Predicted output : \n", output)
```

# 9. Construct AND Gate and OR Gate using Perceptron


```python
# AND

def activation(out, threshold):
    if out >= threshold:
        return 1
    else:
      return 0

def preceptron(and_input):
    a = [0,0,1,1]
    b = [0,1,0,1]
    y = [0,0,0,1]
    w = [1.4,1.5]
    threshold = 1
    learning_rate = 0.1
    i = 0
    print("Perceptron Training: ")
    print("######################")
    print("--------------------")
    while i<4:
        summation = a[i]*w[0] + b[i]*w[1]
        o = activation(summation, threshold)
        print("Input: "+str(a[i]) + ", "+str(b[i]))
        print("Weight: "+str(w[0]) + ", "+str(w[1]))
        print("Summation: "+str(summation) + " Threshold "+str(threshold))
        print("Actual Output: "+str(y[i])+" Predicted Output: "+str(o))
        if(o!=y[i]):
              print("_\nUpdating Weights")
              w[0]=w[0]+learning_rate*(y[i]-o)*a[i]
              w[1]=w[1]+learning_rate*(y[i]-o)*b[i]
              print("Updating Weights: "+str(w[0]) + ", "+str(w[1]))
              i = -1
              print("\nWeights updated Traning Again: ")
              print("##################################")
        i = i + 1
        print("--------------------------")
    summation = and_input[0]*w[0] + and_input[1]*w[1]
    return activation(summation, threshold)
and_input = [1,1]
print("AND Gate Output for "+str(and_input)+" : "+str(preceptron(and_input)))
```

# 10. Working of Perceptron.


```python
import numpy as np

def activation(x):
    return 1 if x >= 0 else 0

def predict(inputs, weights, bias):
    weighted_sum = np.dot(inputs, weights)+bias
    return activation(weighted_sum)


def perceptron(training_inputs, lables, learning_rate = 0.01, num_epochs = 100):
    num_inputs = len(training_inputs[0])
    weights = np.random.rand(num_inputs)
    bias = np.random.rand()

    for epoch in range(num_epochs):
        for inputs, label in zip(training_inputs,labels):
            prediction = predict(inputs, weights, bias)
            error = label - prediction
            weights += learning_rate * error * inputs
            bias += learning_rate * error
    return weights, bias


if __name__ == "__main__":
    training_inputs = np.array(([[0,0], [0,1], [1,0], [1,1]]))
    labels = np.array([0,0,0,1])
    weights, bias = perceptron(training_inputs, labels)
    test_inputs = np.array([[0,0], [0,1], [1,0], [1,1]])
    for inputs in test_inputs:
        print(f"Input: {inputs}, Predicted Output: {predict(inputs, weights, bias)}")
```
